```python
null_values = data.isnull().sum()

# Display the columns with null values and their counts
print("Columns with Null Values:")
print(null_values[null_values > 0])
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# # Create a list of lists representing the transactions
# transactions = data.values.tolist()
```

```python
columns_for_transactions = ["Venue", "Win/Lost", "Batsman", "Opponent Team", "Runs Scored Range", "Strike Rate Range", "Balls Faced Range"]
# Convert the "Batsman" column to a list of lists
batsmen_transactions = data["Batsman"].apply(lambda x: [x]).tolist()

# Combine the "Batsman" transactions with the rest of the columns
transactions = [row[1:] + batsman for row, batsman in zip(data[columns_for_transactions].values.tolist(), batsmen_transactions)]

# Convert the transactions into a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_encoded, min_support=0.005, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Display the frequent itemsets and association rules
frequent_itemsets
# print("\nAssociation Rules:")
# print(rules)
```
