```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# # Create a list of lists representing the transactions
# transactions = data.values.tolist()
```

```python
columns_for_transactions = ["Venue", "Win/Lost", "Batsman", "Opponent Team", "Runs Scored Range", "Strike Rate Range", "Innings","Batting Order"]
# Convert the "Batsman" column to a list of lists
batsmen_transactions = PAK_batsman_df["Batsman"].apply(lambda x: [x]).tolist()

# Combine the "Batsman" transactions with the rest of the columns
transactions = [row[1:] + batsman for row, batsman in zip(PAK_batsman_df[columns_for_transactions].values.tolist(), batsmen_transactions)]

# Convert the transactions into a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_encoded, min_support=0.005, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Display the frequent itemsets and association rules
frequent_itemsets
# print("\nAssociation Rules:")
# print(rules)
```

```python
dataset_size = 205  ## number of matches in the dataset
def calculate_kulczynski_measure(rules):
    precision = rules['confidence']
    recall = rules['consequent support']
    rules['kulczynski_measure'] = (precision + recall) / 2
    return rules['kulczynski_measure']

def calculate_jaccard_coefficient(rules):
    intersection_size = rules['support'] * dataset_size
    union_size = (rules['antecedent support'] * dataset_size) + (rules['consequent support'] * dataset_size) - intersection_size
    rules['jaccard_coefficient'] = intersection_size / union_size
    return rules['jaccard_coefficient']
```

```python
rules['kulczynski measure'] = calculate_kulczynski_measure(rules)
rules['jaccard coefficient'] = calculate_jaccard_coefficient(rules)
```

```python
filtered_rules_Lost_antecedents = rules[rules["antecedents"].apply(lambda x: "Lost" in x)]
filtered_rules_Won_antecedents = rules[rules["antecedents"].apply(lambda x: "Won" in x)]
```

```python
# First, normalize the columns. Here we'll use min-max normalization.
for column in ['confidence', 'lift', 'jaccard_coefficient', 'kulczynski_measure']:
    filtered_rules_Won[column + '_norm'] = (filtered_rules_Won[column] - filtered_rules_Won[column].min()) / (filtered_rules_Won[column].max() - filtered_rules_Won[column].min())

# Assign weights to the attributes based on their importance.
weights = {
    'confidence_norm': 0.30,
    'lift_norm': 0.10,
    'jaccard_coefficient_norm': 0.30,
    'kulczynski_measure_norm': 0.30
}

# Calculate the weighted score for each association rule.
filtered_rules_Won['score'] = sum(filtered_rules_Won[attribute] * weight for attribute, weight in weights.items())

# Sort the DataFrame based on the score in descending order to rank the rules.
filtered_rules_Won = filtered_rules_Won.sort_values(by='score', ascending=False)

#association rules ranked by their score.
```
