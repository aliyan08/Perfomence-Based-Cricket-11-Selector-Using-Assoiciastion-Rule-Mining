### Pakistan Team Batsman's Statistics

```python
df["Batter"]
```

```python
Batsman_Matches_Played = df.groupby('Batter')['Match ID'].nunique()
Batsman_Matches_Played_df = Batsman_Matches_Played.reset_index()
Batsman_Matches_Played_df.rename(columns={'Match ID': 'Matches_played'}, inplace=True)

Batsman_Matches_Played_df.sort_values(by="Matches_played", ascending=False)
```

```python
# batter when Pakistan bats first and Innings 1
list(df[(df['Bat First'] == 'Pakistan') & (df['Innings'] == 1)]["Batter"])
# batter when Pakistan bats second and Innings 2
list(df[(df['Bat Second'] == 'Pakistan') & (df['Innings'] == 2)]["Batter"])

# merge both lists
list_PAK_Batsmans=list(set(list(df[(df['Bat First'] == 'Pakistan') & (df['Innings'] == 1)]["Batter"]) + list(df[(df['Bat Second'] == 'Pakistan') & (df['Innings'] == 2)]["Batter"])))
```

```python
list_PAK_Batsmans
```

```python
len(list_PAK_Batsmans)
```

```python
# matcher played by PAK batsman
matches_by_PAK_Batsman = Batsman_Matches_Played_df[Batsman_Matches_Played_df["Batter"].isin(list_PAK_Batsmans)]
matches_by_PAK_Batsman=matches_by_PAK_Batsman.sort_values(by="Matches_played", ascending=False)
```

```python
matches_by_PAK_Batsman
```

```python
# Total Runs By Pak Batsman
Total_Runs_by_PAK_Batsman = df.groupby('Batter')['Batter Runs'].sum()
Total_Runs_by_PAK_Batsman_df = Total_Runs_by_PAK_Batsman.reset_index()
Total_Runs_by_PAK_Batsman_df.rename(columns={'Batter Runs': 'Total Runs'}, inplace=True)
Total_Runs_by_PAK_Batsman_df = Total_Runs_by_PAK_Batsman_df[Total_Runs_by_PAK_Batsman_df["Batter"].isin(list_PAK_Batsmans)]
Total_Runs_by_PAK_Batsman_df.sort_values(by="Total Runs", ascending=False)
```

```python
# Total Ball Faced By each PAK batsman
Total_Ball_Faced_by_PAK_Batsman = df.groupby('Batter')['Ball'].count()
Total_Ball_Faced_by_PAK_Batsman_df = Total_Ball_Faced_by_PAK_Batsman.reset_index()
Total_Ball_Faced_by_PAK_Batsman_df.rename(columns={'Ball': 'Total Ball Faced'}, inplace=True)
Total_Ball_Faced_by_PAK_Batsman_df = Total_Ball_Faced_by_PAK_Batsman_df[Total_Ball_Faced_by_PAK_Batsman_df["Batter"].isin(list_PAK_Batsmans)]
Total_Ball_Faced_by_PAK_Batsman_df.sort_values(by="Total Ball Faced", ascending=False)
```

```python
# Calculate How Much Time batsman got out
player_out=dict(df['Player Out'].value_counts())
player_out_df = pd.DataFrame.from_dict(player_out, orient='index').reset_index()
player_out_df.rename(columns={'index': 'Batter', 'Player Out': 'Total'}, inplace=True)
player_out_df.rename(columns={ 0 : 'Number_Times_Out'}, inplace=True)
PAK_player_out_df = player_out_df[player_out_df["Batter"].isin(list_PAK_Batsmans)]
PAK_player_out_df.sort_values(by="Number_Times_Out", ascending=False)
```

```python
# Find players who are not in PAK_player_out_df
players_not_in_df = set(list_PAK_Batsmans) - set(PAK_player_out_df["Batter"])

# Create a DataFrame for missing players and set the number of times out to 0
missing_players_df = pd.DataFrame({"Batter": list(players_not_in_df), "Number_Times_Out": 0})

# Append missing players to PAK_player_out_df
PAK_player_out_df = pd.concat([PAK_player_out_df, missing_players_df], ignore_index=True, sort=False)
PAK_player_out_df = PAK_player_out_df.sort_values(by="Number_Times_Out", ascending=False).reset_index(drop=True)
PAK_player_out_df
```

```python
# Merge All  DataFrames on Batter
PAK_Batsman_Stats_df = pd.merge(Total_Runs_by_PAK_Batsman_df, Total_Ball_Faced_by_PAK_Batsman_df, on='Batter')
PAK_Batsman_Stats_df = pd.merge(PAK_Batsman_Stats_df, matches_by_PAK_Batsman, on='Batter')
PAK_Batsman_Stats_df = pd.merge(PAK_Batsman_Stats_df, PAK_player_out_df, on='Batter')
PAK_Batsman_Stats_df
```

```python
PAK_Batsman_Stats_df["Scoring AVG"] = round(PAK_Batsman_Stats_df["Total Runs"] / PAK_Batsman_Stats_df["Number_Times_Out"], 2)
PAK_Batsman_Stats_df["STRIKE_RATE"] = round((PAK_Batsman_Stats_df["Total Runs"] / PAK_Batsman_Stats_df["Total Ball Faced"]) * 100, 2)
PAK_Batsman_Stats_df
```

### Considering Pakistan Matches Only

```python
pakistan_matches_df = df[(df['Bat First'] == 'Pakistan') | (df['Bat Second'] == 'Pakistan')]
pakistan_matches_df
```

```python
batsman_runs_df = pd.DataFrame(columns=['Match ID','Date', 'Venue', 'Winner', 'Batsman', 'Runs Scored', 'Balls Faced', 'Opponent Team'])

# Group the original DataFrame by 'Match ID' and 'Batter'
grouped_data = pakistan_matches_df.groupby(['Match ID', 'Batter'])

# Iterate through each group
for (match_id, batsman), group_data in grouped_data:
    total_runs = group_data['Batter Runs'].sum()
    
    # Count only valid balls (where 'Valid Ball' column is 1)
    valid_balls = group_data['Valid Ball'] == 1
    total_balls_faced = valid_balls.sum()
    
    inning = group_data['Innings'].iloc[0]
    team_batsman_belongs_to = group_data['Bat First'].iloc[0] if inning == 1 else group_data['Bat Second'].iloc[0]
    
    # Determine the opponent team based on the inning
    opponent_team = group_data['Bat Second'].iloc[0] if inning == 1 else group_data['Bat First'].iloc[0]

    # Get additional match details
    match_details = df[df['Match ID'] == match_id].iloc[0]
    date = match_details['Date']
    venue = match_details['Venue']
    winner = match_details['Winner']

    # Append data to the new DataFrame
    batsman_runs_df = pd.concat([batsman_runs_df, pd.DataFrame({
        'Match ID' : [match_id],
        'Date': [date],
        'Venue': [venue],
        'Winner': [winner],
        'Batsman': [batsman],
        'Runs Scored': [total_runs],
        'Balls Faced': [total_balls_faced],
        'Opponent Team': [opponent_team]
    })], ignore_index=True)

# Display the resulting DataFrame
    
# Display the resulting DataFrame
batsman_runs_df
```

#### Calculating batting positions of Batsman in each match

```python
# itrate through rows in batsman_runs_df
batting_position_df = pd.DataFrame(columns=['Match ID','Date','Venue','Winner','Batsman','Batting Position', 'Runs Scored', 'Balls Faced', 'Opponent Team'])
buffer_matchID = 0
batting_position = 0
for index, row in batsman_runs_df.iterrows():
    
    if row['Match ID'] != buffer_matchID:
        batting_position = 0
        buffer_matchID = row['Match ID']
    
    # get the batting position
    batting_position +=1
    
    # get the match id
    match_id = row['Match ID']
    # get the date
    date = row['Date']
    # get the venue
    venue = row['Venue']
    # get the winner
    winner = row['Winner']
    # get the batsman
    batsman = row['Batsman']
     
    # get the runs scored
    runs_scored = row['Runs Scored']
    # get the balls faced
    balls_faced = row['Balls Faced']
    # get the opponent team
    opponent_team = row['Opponent Team']
    
    # append the data to the new DataFrame
    batting_position_df = pd.concat([batting_position_df, pd.DataFrame({
        'Match ID' : [match_id],
        'Date': [date],
        'Venue': [venue],
        'Winner': [winner],
        'Batsman': [batsman],
        'Batting Position': [batting_position],
        'Runs Scored': [runs_scored],
        'Balls Faced': [balls_faced],
        'Opponent Team': [opponent_team]
    })], ignore_index=True)

batting_position_df
```

```python
PAK_Batsman_Matches_df = batting_position_df
```

```python
# funtion to calculate strike rate
def calculate_strike_rate(runs_scored, balls_faced):
    if balls_faced == 0:
        return 0
    else:
        return round(runs_scored / balls_faced * 100, 2)
```

```python
PAK_Batsman_Matches_df["Strike Rate"] = PAK_Batsman_Matches_df.apply(lambda row: calculate_strike_rate(row["Runs Scored"], row["Balls Faced"]), axis=1)
PAK_Batsman_Matches_df
```

```python
def runs_scored_inrange(runs_scored):
    if runs_scored < 25:
        return "Runs < 25"
    elif runs_scored >= 25 and runs_scored < 50:
        return "Runs >= 25 and < 50"
    elif runs_scored >= 50:
        return "Runs >= 50"

def StrikeRate_inrange(strike_rate):
    if strike_rate < 90:
        return "S/R < 90"
    elif strike_rate >= 90 and strike_rate < 120:
        return "S/R >= 90 and < 120"
    elif strike_rate >= 120:
        return "S/R >= 120"

def balls_faced_inrange(balls_faced):
    if balls_faced < 20:
        return "B/F < 20"
    elif balls_faced >= 20 and balls_faced < 40:
        return "B/F >= 20 and <= 40"
    elif balls_faced >= 40:
        return "B/F >= 40"
```

```python
Final_PAK_Batsman_Matches_df["Runs Scored Range"] = Final_PAK_Batsman_Matches_df.apply(lambda row: runs_scored_inrange(row["Runs Scored"]), axis=1)
Final_PAK_Batsman_Matches_df["Strike Rate Range"] = Final_PAK_Batsman_Matches_df.apply(lambda row: StrikeRate_inrange(row["Strike Rate"]), axis=1)
Final_PAK_Batsman_Matches_df["Balls Faced Range"] = Final_PAK_Batsman_Matches_df.apply(lambda row: balls_faced_inrange(row["Balls Faced"]), axis=1)
Final_PAK_Batsman_Matches_df
```

```python
Final_PAK_Batsman_Matches_df.drop(columns=["Runs Scored", "Strike Rate", "Balls Faced"], inplace=True)
Final_PAK_Batsman_Matches_df
```
