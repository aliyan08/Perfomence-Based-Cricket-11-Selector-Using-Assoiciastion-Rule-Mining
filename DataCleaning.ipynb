#### Calculating batting positions of Batsman in each match

```python
import pandas as pd

# Assuming 'pakistan_matches_df' is your dataframe with the mentioned columns.

# Initialize a dictionary to hold batting positions for each match.
batting_positions = {}

# Iterate over the DataFrame sorted by 'Match ID', 'Innings', and delivery ('Over' and 'Ball').
for idx, delivery in pakistan_matches_df.sort_values(by=['Match ID', 'Innings', 'Over', 'Ball']).iterrows():
    match_id = delivery['Match ID']
    innings = delivery['Innings']
    batsman = delivery['Batter']
    non_striker = delivery['Non Striker']
    player_out = delivery['Wicket']
    
    # Initialize the match and innings in the dictionary if not present.
    if (match_id, innings) not in batting_positions:
        batting_positions[(match_id, innings)] = {'batsmen': [], 'positions': {}}
    
    # For the first two batsmen.
    if not batting_positions[(match_id, innings)]['batsmen']:
        batting_positions[(match_id, innings)]['batsmen'].extend([batsman, non_striker])
        batting_positions[(match_id, innings)]['positions'][batsman] = 1
        batting_positions[(match_id, innings)]['positions'][non_striker] = 2
    
    # If a wicket falls, assign the next position to the new batsman.
    if player_out == 1:
        player_dismissed = delivery['Batter'] if delivery['Batter'] == delivery['Player Out'] else delivery['Non Striker']
        # Find the new batsman who is not in the existing 'batsmen' list.
        new_batsman = batsman if batsman not in batting_positions[(match_id, innings)]['batsmen'] else non_striker
        next_position = len(batting_positions[(match_id, innings)]['batsmen']) + 1
        batting_positions[(match_id, innings)]['batsmen'].append(new_batsman)
        batting_positions[(match_id, innings)]['positions'][new_batsman] = next_position

# Convert the dictionary to a DataFrame for easy manipulation and viewing.
batting_positions_df = pd.DataFrame([(match_id, innings, batsman, position) 
                                     for (match_id, innings), data in batting_positions.items() 
                                     for batsman, position in data['positions'].items()],
                                    columns=['Match ID', 'Innings', 'Batsman', 'Batting Position'])

# Merging the batting position data with the original batsman_runs_df to include batting positions.
batsman_runs_with_positions_df = batsman_runs_df.merge(batting_positions_df, on=['Match ID', 'Batsman'], how='left')

# Display the DataFrame with batting positions.
batsman_runs_with_positions_df
```

```python
def runs_scored_inrange(runs_scored):
    if runs_scored < 25:
        return "Runs < 25"
    elif runs_scored >= 25 and runs_scored < 50:
        return "Runs >= 25 and < 50"
    elif runs_scored >= 50:
        return "Runs >= 50"

def StrikeRate_inrange(strike_rate):
    if strike_rate < 90:
        return "S/R < 90"
    elif strike_rate >= 90 and strike_rate < 120:
        return "S/R >= 90 and < 120"
    elif strike_rate >= 120:
        return "S/R >= 120"

def balls_faced_inrange(balls_faced):
    if balls_faced < 20:
        return "B/F < 20"
    elif balls_faced >= 20 and balls_faced < 40:
        return "B/F >= 20 and <= 40"
    elif balls_faced >= 40:
        return "B/F >= 40"
def calculate_innings(innings):
    if innings == 1:
        return "1st Inning"
    else:
        return "2nd Inning"

def calculate_batting_position(batting_position):
    if batting_position <= 3:
        return "Top Order"
    elif batting_position > 3 and batting_position <= 6:
        return "Middle Order"
    elif batting_position > 6:
        return "Low Order"
```

```python
Final_PAK_Batsman_Matches_df["Runs Scored Range"] = Final_PAK_Batsman_Matches_df.apply(lambda row: runs_scored_inrange(row["Runs Scored"]), axis=1)
Final_PAK_Batsman_Matches_df["Strike Rate Range"] = Final_PAK_Batsman_Matches_df.apply(lambda row: StrikeRate_inrange(row["Strike Rate"]), axis=1)
Final_PAK_Batsman_Matches_df["Balls Faced Range"] = Final_PAK_Batsman_Matches_df.apply(lambda row: balls_faced_inrange(row["Balls Faced"]), axis=1)
Final_PAK_Batsman_Matches_df["Batting Order"] = Final_PAK_Batsman_Matches_df.apply(lambda row: calculate_batting_position(row["Batting Position"]), axis=1)
Final_PAK_Batsman_Matches_df["Innings"] = Final_PAK_Batsman_Matches_df.apply(lambda row: calculate_innings(row["Innings"]), axis=1)
Final_PAK_Batsman_Matches_df
```

```python
bowler_stats_df = pd.DataFrame(columns=['Match ID','Date', 'Venue', 'Winner', 'Bowler', 'Runs Concided','Wickets', 'Economy','Ball Bowled', 'Opponent Team'])

# Group the original DataFrame by 'Match ID' and 'Bowler'
grouped_data = pakistan_matches_df.groupby(['Match ID', 'Bowler'])

# Iterate through each group
for (match_id, bowler), group_data in grouped_data:
    total_runs = group_data['Bowler Runs Conceded'].sum()
    
    valid_balls = group_data['Valid Ball']
    total_balls_bowled = valid_balls.count()
    
    inning = group_data['Innings'].iloc[0]
    team_bowler_belongs_to = group_data['Bat Second'].iloc[0] if inning == 1 else group_data['Bat First'].iloc[0]
    
    # Determine the opponent team based on the inning
    
    opponent_team = group_data['Bat First'].iloc[0] if inning == 1 else group_data['Bat Second'].iloc[0]
    Wickets = group_data['Wicket'].sum()
    # Get additional match details
    match_details = pakistan_matches_df[pakistan_matches_df['Match ID'] == match_id].iloc[0]
    date = match_details['Date']
    venue = match_details['Venue']
    winner = match_details['Winner']

    # Append data to the new DataFrame
    bowler_stats_df = pd.concat([bowler_stats_df, pd.DataFrame({
        'Match ID' : [match_id],
        'Date': [date],
        'Venue': [venue],
        'Winner': [winner],
        'Bowler': [bowler],
        'Runs Concided': [total_runs],
        'Wickets': [Wickets],
        'Ball Bowled': [total_balls_bowled],
        'Opponent Team': [opponent_team]
    })], ignore_index=True)

# Display the resulting DataFrame
bowler_stats_df
```

```python
def calculate_economy(balls, runs):
    if balls % 6 == 0:
        overs = balls // 6
        return round(runs / overs, 2) if overs > 0 else runs
    else:
        balls = balls - (balls % 6)
        overs = balls // 6
        return round(runs / overs, 2) if overs > 0 else runs

def calculate_over_bowled(balls):
    if balls % 6 == 0:
        overs = balls // 6
        return overs if overs > 0 else 1 
    else:
        balls = balls - (balls % 6)
        overs = balls // 6
        return overs if overs > 0 else 1
```
